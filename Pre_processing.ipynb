```python
threat_zone = 'Zone17'
number_obs_zones = 8
input_tz = 'TZ17/tz17_combined.csv'
input_1 = 'stage1_sample_submission.csv'
subm = 'Submissions/stage1_submission_fileZone16.csv'
df = pd.read_csv(input_tz)
sub_list = pd.read_csv(input_1)
sub_file = pd.read_csv(subm)
#df.iloc[0,1]
```

```python
df.columns = ['0', 'Probability']
df
```

```python
def assort(df, sub_list, threat_zone, number_obs_zones):
    df = df.groupby(np.arange(len(df.index))//number_obs_zones).mean()
    sub_list['Subject'], sub_list['Zone'] = sub_list['Id'].str.split('_',1).str
    sub_list = sub_list['Subject'].unique() + '_' + threat_zone
    #sub_list = sub_list.tolist()
    sub_list = pd.DataFrame(sub_list)
    sub_list.columns = ['Subject']
    df['Id'] = sub_list['Subject']
    df = df[['Id', '0', 'Probability']]
    return df
```

```python
df = assort(df, sub_list, threat_zone, number_obs_zones)
df
```

```python
merged_df = sub_file.merge(df, how='left', left_on='Id', right_on='Id')
merged_df['Probability_y'] =  merged_df['Probability_y'].replace(np.nan,  merged_df['Probability_x'])
    
merged_df
```

```python
merged_df.drop('Probability_x', axis=1, inplace=True)
merged_df.drop('0', axis=1, inplace=True)
merged_df.columns = ['Id', 'Probability']
merged_df 
```

```python
f_name = 'Submissions/stage1_submission_file' + threat_zone + '.csv'
f_name
```

```python
merged_df.to_csv(f_name, index=False)
```
